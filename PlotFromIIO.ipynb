```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
filename = 'test.iq'
sample_rate = 2e6
```

```python
with open(filename, 'rb') as f:
    data = f.read()
```

```python
data_array = np.frombuffer(data, dtype='<i2')
```

```python
data_iq = data_array[::2] + 1j * data_array[1::2]
```

```python
plt.figure()
plt.xlabel("Time")
plt.ylabel("Frequency (Hz)")
Pxx, freqs, bins, im = plt.gca().specgram(data_iq, NFFT=256, Fs=sample_rate, noverlap=10)
plt.tight_layout()
```
