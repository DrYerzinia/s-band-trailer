```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget

import pickle
```

```python
name = 'scan_c_'
```

```python
az_meas = np.load(name+'az_meas.npy')
el_meas = np.load(name+'el_meas.npy')
raw_samples = np.load(name+'raw_samples.npy')
```

```python
pwrs = np.zeros(raw_samples.shape[0])
for i in range(az_meas.shape[0]):
    pwrs[i] = 10*np.log10(np.max(np.abs(np.fft.fft(raw_samples[i]))[3900:4050])**2)
```

```python
plt.figure()
shift = 3
plt.tripcolor(az_meas[:-shift], el_meas[:-shift], pwrs[shift:], shading='gouraud')
#plt.scatter(az_meas, el_meas, c=pwrs)
plt.colorbar()
plt.axis('equal')
plt.grid()
```
