```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget

import pickle
```

```python
names = ['scan_a_', 'scan_b_', 'scan_c_']
```

```python
plt.figure()

max = 0
for name in names:
    az_meas = np.load(name+'az_meas.npy')
    el_meas = np.load(name+'el_meas.npy')
    raw_samples = np.load(name+'raw_samples.npy')

    pwrs = np.zeros(raw_samples.shape[0])
    for i in range(raw_samples.shape[0]):
        #pwrs[i] = 10*np.log10(np.max(np.abs(np.fft.fft(raw_samples[i]))[3900:4050])**2)
        pwrs[i] = np.max(np.abs(np.fft.fft(raw_samples[i]))[3900:4050])
    if np.max(pwrs) > max:
        max = np.max(pwrs)
    print(np.max(pwrs))
    print(np.argmax(pwrs))

    #plt.tripcolor(az_meas, el_meas, pwrs, vmin=0, vmax=max)#, shading='gouraud')
    plt.scatter(az_meas, el_meas, c=pwrs, vmin=0, vmax=max)#, shading='gouraud')

plt.colorbar()
plt.axis('equal')
plt.grid()
```
